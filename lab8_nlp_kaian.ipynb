Natural language processing (NLP) is an interdisciplinary subfield of computer science and linguistics. It is primarily concerned with giving computers the ability to support and manipulate speech. It involves processing natural language datasets, such as text corpora or speech corpora, using either rule-based or probabilistic (i.e. statistical and, most recently, neural network-based) machine learning approaches. The goal is a computer capable of "understanding" the contents of documents (excerpted from [Wikipedia](https://en.wikipedia.org/wiki/Natural_language_processing)). In this lab, we learn some basic techniques of processing textual data.

First, we need to define a document which is a string for NLP analysis:

```python
# Define a document containing a description or text
document = "The programme aims to enlighten students in developing their competence in identifying, learning and applying emerging educational technologies, pilot and develop technology enhanced teaching and learning strategies in both online and offline learning environments. Graduates will be capable for demonstrating their skills in educational technologies and ready for different positions in educational institutions as teaching staff, learning designer, and senior IT-related administrative member."
```

```python
# Print the length of the document, which represents the number of characters in the text
print(len(document)) # 字元

# Print the content of the document, which is the text itself
print(document)
```

## 1. Use the NLTK Package

We need NLP packages to help us to analyse the text. [NLTK](https://www.nltk.org)<sup>*</sup> (Natural Language Toolkit) is a famous package for processing English text. Below, we import this package for our use. It needs to download additional data when we use it for the first time.

\* NLTK was developed by Prof Steven Bird (my former colleague at the University of Melbourne)



```python
# Import the Natural Language Toolkit (nltk) library
import nltk

# Download the 'punkt' dataset, which includes tokenizers
nltk.download('punkt')

# Download the 'stopwords' dataset, which includes common stop words
nltk.download('stopwords')

# Download extra datasets, which is a lexical database for the English language
nltk.download('wordnet')
nltk.download('omw-1.4')
```

## 2. Case Folding

The following code examples show how to change the case of the document. Note that you can either ``lower`` or ``upper`` the sentence, depending on what you want to do next.

```python
# Convert the text in 'document' to lowercase
document_lower = document.lower()

# Print the text in lowercase
print(document_lower)
```

```python
# Convert the text in 'document' to uppercase
document_upper = document.upper()

# Print the text in uppercase
print(document_upper)
```

## 3. Word Tokenization

Word tokenization is the process of breaking a text into individual words, phrases, symbols, or other meaningful elements, referred to as tokens. Essentially, tokenization helps in segmenting text into meaningful units, which could be words, punctuation marks, or any other significant parts of the text. Our goal is to split the sentences apart so that we can more do further research.

```python
# Import the 'word_tokenize' function from the Natural Language Toolkit (nltk)
from nltk import word_tokenize

# Tokenize the lowercase text in 'document_lower' into individual words
words_nltk = word_tokenize(document_lower)

# Iterate through each token (word) in the list of tokens (the first 25 only)
for word in words_nltk[:25]:
    # Print each word on a separate line
    print(word)
```

## 4. Sentence Segmentation

Sentence segmentation, also known as sentence boundary disambiguation, is the process of identifying the boundaries in a text that separate sentences. In simpler terms, it's like figuring out where one sentence ends and another begins in a paragraph or a larger body of text. This is an essential step in NLP too because it helps computers understand the structure of written language.

```python
# Import the 'sent_tokenize' function from the Natural Language Toolkit (nltk)
from nltk import sent_tokenize

# Tokenize the lowercase text in 'document_lower' into individual sentences
sentences_nltk = sent_tokenize(document_lower)

# Iterate through each sentence in the list of sentences
for sentence in sentences_nltk:
    # Print each sentence on a separate line
    print(sentence)
```

## 5. Removing Punctuations

Removing punctuations refers to the process of eliminating punctuation marks from a piece of text.

Punctuation marks include symbols such as periods, commas, question marks, exclamation points, and others that are used to enhance the structure and meaning of sentences. However, these are not very useful for our data analysis.

```python
# Import the 'punctuation' string from the 'string' module
from string import punctuation

# Print the 'punctuation' string, which contains common punctuation characters
print(punctuation)
```

```python
# Show the "words" that are not punctuations:
for word in words_nltk[:25]:
    if word not in punctuation:
        print(word)
```

Now, we try to use a for-loop to count how many times that each word appears in the data. This is the most basic form of NLP analysis called **word frequencies analysis**.

```python
from collections import Counter
```

```python
# Initialize an empty dictionary to store word frequencies
word_freq = Counter()

# Tokenize the lowercase text in 'document_lower' using nltk's word_tokenize
words_nltk = word_tokenize(document_lower)

# Iterate through each word in the tokenized text
for word in words_nltk:
    # Check if the word is not a punctuation character
    if word not in punctuation:
        # Check if the word is not already in the word frequency dictionary
        if word not in word_freq:
            # If it's not in the dictionary, add it with a frequency of 1
            word_freq[word] = 1
        else:
            # If it's already in the dictionary, increment its frequency
            word_freq[word] += 1

# Iterate through the sorted word frequencies and print each word and its frequency
for freq in word_freq.most_common(20):
    print(freq)

# Print the total number of tokens (unique words) in the text
print('total number of tokens', len(word_freq))
```

## 6. Removing *Stopwords*

**Stopwords** are common words that are often removed from text during the process of NLP because they do not carry significant meaning. These words include articles (like "a," "an," "the"), prepositions (like "in," "on," "at"), conjunctions (like "and," "but," "or"), and other frequently used words (like "is," "are," "have").

The reason for removing stopwords is to focus on the more important words in a document or a piece of text. By eliminating these common words, the analysis can concentrate on the essential content-carrying words, which can lead to more effective processing in tasks such as text classification, sentiment analysis, and information retrieval.

```python
# Import the 'stopwords' corpus from NLTK
from nltk.corpus import stopwords

# Retrieve the set of English stopwords using the 'stopwords.words' function
stopword_list = set(stopwords.words('english'))

# Print the number of stopwords in the list
print('number of stopword list: ', len(stopword_list))

print(f'stopword_list: \n{stopword_list}')
```

```python
# Initialize an empty dictionary to store word frequencies
word_freq = Counter()

# Tokenize the lowercase text in 'document_lower' using nltk's word_tokenize
words_nltk = word_tokenize(document_lower)

# Iterate through each word in the tokenized text
for word in words_nltk:
    # Check if the word is not a punctuation character and not in the stopword list
    if (word not in punctuation) and (word not in stopword_list):
        # Check if the word is not already in the word frequency dictionary
        if word not in word_freq:
            # If it's not in the dictionary, add it with a frequency of 1
            word_freq[word] = 1
        else:
            # If it's already in the dictionary, increment its frequency
            word_freq[word] += 1

# Iterate through the sorted word frequencies and print each word and its frequency
for freq in word_freq.most_common(20):
    print(freq)

# Print the total number of tokens (unique words) in the text
print('total number of tokens', len(word_freq))
```

## 7. Stemming and Lemmatization

Both stemming and lemmatization are used to standardize words, making it easier for NLP systems to process and analyze text by treating different forms of the same word as one. While stemming is faster and more straightforward, lemmatization produces more accurate results at the cost of increased computational complexity.

```python
# Import the Porter Stemmer from the Natural Language Toolkit (nltk.stem)
from nltk.stem import PorterStemmer

# Create an instance of the Porter Stemmer
stemmer = PorterStemmer()

# Define a list of example words to be stemmed
example_words = ['demonstrate', 'demonstration', 'demonstrating']

# Iterate through each example word
for example_word in example_words:
    # Print the original word and its stemmed form
    print(example_word, 'is stemmed to', stemmer.stem(example_word))
```

```python
# Import the WordNetLemmatizer from the Natural Language Toolkit (nltk.stem)
from nltk.stem import WordNetLemmatizer

# Create an instance of the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Define a list of example words to be lemmatized
example_words = ['demonstrate', 'demonstration', 'demonstrating']

# Iterate through each example word
for example_word in example_words:
    # Print the original word and its lemmatized form
    print(example_word, 'is lemmatized to', lemmatizer.lemmatize(example_word, "v"))
```

To start with a simple example, we use stemming in this case:

```python
# Initialize an empty dictionary to store word frequencies
word_freq = Counter()

# Tokenize the lowercase text in 'document_lower' using nltk's word_tokenize
words_nltk = word_tokenize(document_lower)

# Iterate through each word in the tokenized text
for word in words_nltk:
    # Check if the word is not a punctuation character and not in the stopword list
    if word not in punctuation and word not in stopword_list:
        # Apply stemming to the word using the Porter Stemmer
        word = stemmer.stem(word)

        # Check if the stemmed word is not already in the word frequency dictionary
        if word not in word_freq:
            # If it's not in the dictionary, add it with a frequency of 1
            word_freq[word] = 1
        else:
            # If it's already in the dictionary, increment its frequency
            word_freq[word] += 1

# Iterate through the sorted word frequencies and print each word and its frequency
for freq in word_freq.most_common(20):
    print(freq)

# Print the total number of tokens (unique words) in the text
print('total number of tokens', len(word_freq))
```
